# Final Project Write Up and EDA

### Video games are a $99.6 billion dollar industry with steep monetary and human development costs. With the price of each unit not changing since 2006 and the cost of development skyrocketing yearly sales trends are incredibly valuable. Figuring out which kinds of games will sell millions of copies is paramount for publishers and developers to stay in business.

## Potential Methods

1) Logistic regression deciding whether or not to invest in a game based on the variables at hand.
- Create a binary variable based on a decided sales target to predict against

2) Clustering to see if there are characteristics that define a successful game based on our data.
-  We may not have enough variables for this approach

Variable | Description | Type of Variable
---|---|---
Name | Name of the game | Categorical
Platform | Console game released on | Categorical
Year of Release | Year game released | Discrete
Genre | Genre | Categorical
Publisher | Publishing company | Categorical
NA_Sales | Sales in North America (in millions) | Continuous
EU_Sales | Sales in Europe (in millions) | Continuous
JP_Sales | Sales in Japan (in millions) | Continuous
Other_Sales | Sales in the rest of the world (in millions) | Countinuous
Global_Sales | Total sales (in millions) | Continuous
Critic_score | Aggregate score compiled by Metacritic staff | Continuous
Critic_count | The number of critics used in coming up with the Critic_score | Continuous
User_score | Score by Metacritic's subscribers | Continuous
User_count | Number of users who gave the user_score | Continuous
Developer | Party responsible for creating the game | Categorical
Rating | The ESRB ratings | Categorical

## Project Concerns

The largest concern I have is that there is not sufficient data to accurately predict whether a game will be successful or not. There are a lot of nuances lost in the dataset that were either from cleaning the authors performed or were never there from the data scrape. 

There is also significant data that is simply unknown. Unit sales are generally known to the public, but we don't know how many were sold through a sale, deal, or given for free as a bundle. We're forced to assume that the units sold are all equal. This simply is not true because of the rise of digital purchases and sales.

Lastly, units sold is not the only source of revenue that a game can provide. The rise of "Games as a Service" has affected every developer, publisher, and console. High speed internet allows significant post-launch support for many titles ranging from multiplayer to additional downloadable content.



## Outcomes

- I expect Critic Score to have the most impact, sadly. The general public relies on reviews of products to guide purchasing decisions where they aren't an expert
- Complexity shouldn't be too significant. However, there are several variables that would have to be dummied up or left out
- For success, the model needs to be able to predict if a game will sell over X units at least 85% of the time. Market trends can aid decisions, but sometimes good products don't sell
