# Build your models 

```python
from sklearn import linear_model
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import  train_test_split
from sklearn import model_selection
```

```python
X = df.drop('badteeth', axis = 1)
X.drop('Country', axis = 1, inplace = True)
```

```python
y = df['badteeth']
```

```python
train_X, test_X, train_y, test_y = model_selection.train_test_split(X, y, train_size = .6, random_state = 0)
```

```python
X.columns
```

```python
X.corr()
```

```python
model = linear_model.LinearRegression().fit(train_X, train_y)
```

```python
print model.coef_
print model.score(train_X, train_y)
print model.score(test_X, test_y)
```

```python
model = tree.DecisionTreeRegressor(random_state = 0).\
    fit(train_X, train_y)
```

```python
train_y_hat = model.predict(train_X)
print np.sqrt(metrics.mean_squared_error(train_y, train_y_hat))

test_y_hat = model.predict(test_X)
print np.sqrt(metrics.mean_squared_error(test_y, test_y_hat))
```

```python
model = smf.ols(formula = 'badteeth ~ literacy_avg + GDP_avg + sugar_avg + health_avg', data = df).fit()
model.summary()
```

```python
model = smf.ols(formula = 'badteeth ~ literacy_avg + sugar_avg + GDP_avg + health_avg', data = df).fit()
model.summary()
```
